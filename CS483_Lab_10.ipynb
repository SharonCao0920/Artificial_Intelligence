```python
"""
3.Holidays
"""
import pandas as pd
df = pd.read_csv("aapl_no_dates_2.csv") # compare with aapl_3.csv
print(df.head())
print("------------------------------------------------------")
rng = pd.date_range(start="7/1/2017", end="7/21/2017", freq='B')
print(rng)
print("------------------------------------------------------")
from pandas.tseries.holiday import USFederalHolidayCalendar
from pandas.tseries.offsets import CustomBusinessDay

us_cal = CustomBusinessDay(calendar=USFederalHolidayCalendar())

rng = pd.date_range(start="7/1/2017",end="7/23/2017", freq=us_cal)
print(rng)
print("------------------------------------------------------")

df.set_index(rng,inplace=True) 
print(df.head())
print("------------------------------------------------------")
from pandas.tseries.holiday import AbstractHolidayCalendar, nearest_workday, Holiday
class myCalendar(AbstractHolidayCalendar):
    rules = [
        Holiday('My Birth Day', month=4, day=15),#, observance=nearest_workday),
    ]
my_bday = CustomBusinessDay(calendar=myCalendar())
print(pd.date_range('4/1/2017','4/30/2017',freq=my_bday))
print("------------------------------------------------------")
egypt_weekdays = "Sun Mon Tue Wed Thu"
b = CustomBusinessDay(weekmask=egypt_weekdays)
print(pd.date_range(start="7/1/2017",periods=20,freq=b))
print("------------------------------------------------------")
b = CustomBusinessDay(holidays=['2017-07-04', '2017-07-10'], weekmask=egypt_weekdays)
print(pd.date_range(start="7/1/2017",periods=20,freq=b))
```

```python
"""
4.to_datetime: process different date format 
"""
import pandas as pd
dates = ['2017-01-05', 'Jan 5, 2017', '01/05/2017', '2017.01.05', '2017/01/05','20170105']
print(pd.to_datetime(dates))
print("---------------------------------------------------------------------------------")
dt = ['2017-01-05 2:30:00 PM', 'Jan 5, 2017 14:30:00', '01/05/2016', '2017.01.05', '2017/01/05','20170105']
print(pd.to_datetime(dt))
print("---------------------------------------------------------------------------------")
print(pd.to_datetime('30-12-2016'))
print("---------------------------------------------------------------------------------")
print(pd.to_datetime('5-1-2016', dayfirst=True))
print("---------------------------------------------------------------------------------")
print(pd.to_datetime('2017$01$05', format='%Y$%m$%d'))
#	$: delimiter;	format: yy-mm-dd
print("---------------------------------------------------------------------------------")
print(pd.to_datetime('2017#01#05', format='%Y#%m#%d'))
#	#: delimiter;	format: yy-mm-dd
print("---------------------------------------------------------------------------------")
print(pd.to_datetime(['2017-01-05', 'Jan 6, 2017', 'abc'], errors='ignore'))
#	'abc' : intentional error
print("---------------------------------------------------------------------------------")
print(pd.to_datetime(['2017-01-05', 'Jan 6, 2017', 'abc'], errors='coerce'))
print("---------------------------------------------------------------------------------")
# Epoch or Unix time means number of seconds that have passed since Jan 1, 1970 00:00:00 UTC time
# epoch converter: 	https://www.epochconverter.com/
current_epoch = 1594833926
print(pd.to_datetime(current_epoch, unit='s'))
print("---------------------------------------------------------------------------------")
print(pd.to_datetime(current_epoch*1000, unit='ms'))
print("---------------------------------------------------------------------------------")
t = pd.to_datetime([current_epoch], unit='s') 
print(t)
print("---------------------------------------------------------------------------------")
t.view('int64')
```

```python
"""
5.Period and PeriodIndex
"""
import pandas as pd
y = pd.Period('2016')
print(y)		# A-DEC: 	A: annual 
print("---------------------------------------------------------------------------------")
print(y.start_time)
print("---------------------------------------------------------------------------------")
print(y.end_time)
print("---------------------------------------------------------------------------------")
print(y.is_leap_year)
print("---------------------------------------------------------------------------------")
m = pd.Period('2017-12')
print(m)
print("---------------------------------------------------------------------------------")
print(m.start_time)

print("---------------------------------------------------------------------------------")
print(m.end_time)

print("---------------------------------------------------------------------------------")
print(m+1)

print("---------------------------------------------------------------------------------")
d = pd.Period('2016-02-28', freq='D') 
print(d)
print("---------------------------------------------------------------------------------")
print(d.start_time)
print("---------------------------------------------------------------------------------")
print(d.end_time)
print("---------------------------------------------------------------------------------")
print(d+1)
print("---------------------------------------------------------------------------------")
h = pd.Period('2017-08-15 23:00:00',freq='H')
print(h)
print("---------------------------------------------------------------------------------")
print(h+1)
print("---------------------------------------------------------------------------------")
print(h+pd.offsets.Hour(1))	# Achieve same results using pandas offsets hour
print("---------------------------------------------------------------------------------")
q1= pd.Period('2017Q1', freq='Q-JAN')	# Q: quarter
print(q1)
print("---------------------------------------------------------------------------------")

print(q1.start_time)
print("---------------------------------------------------------------------------------")
print(q1.end_time)
print("---------------------------------------------------------------------------------")
print(q1.asfreq('M',how='start'))
print("---------------------------------------------------------------------------------")

print(q1.asfreq('M',how='end'))
print("---------------------------------------------------------------------------------")

w = pd.Period('2017-07-05',freq='W')  # w: week
print(w)
print("---------------------------------------------------------------------------------")
print(w-1)
print("---------------------------------------------------------------------------------")
w2 = pd.Period('2017-08-15',freq='W')
print(w2)
print("---------------------------------------------------------------------------------")
print(w2-w)
print("---------------------------------------------------------------------------------")
r = pd.period_range('2011', '2017', freq='q')
print(r)
print("---------------------------------------------------------------------------------")
print(r[0].start_time)
print("---------------------------------------------------------------------------------")
print(r[0].end_time)
print("---------------------------------------------------------------------------------")
r = pd.period_range('2011', '2017', freq='q-jan')
print(r)
print("---------------------------------------------------------------------------------")
print(r[0].start_time)
print("---------------------------------------------------------------------------------")
print(r[0].end_time)



```

```python
import pandas as pd 
df = pd.read_csv("wmt.csv") 
print(df)
print("---------------------------------------------------------------------------------")
df.set_index("Line Item",inplace=True)
df = df.T		# T: transpose
print(df)
print("---------------------------------------------------------------------------------")
df.index = pd.PeriodIndex(df.index, freq="Q-JAN")
# Change object to "Q-JAN"
print(df)
print("---------------------------------------------------------------------------------")
print(df.index)
print("---------------------------------------------------------------------------------")
print(df.index[0].start_time)
print("---------------------------------------------------------------------------------")
df["Start Date"]=df.index.map(lambda x: x.start_time)
print(df)
print("---------------------------------------------------------------------------------")
df["End Date"]=df.index.map(lambda x: x.end_time)
print(df)
```

```python
"""
6.Timezone Handling
"""
import pandas as pd
df = pd.read_csv("msft.csv", header=1,index_col='Date Time',parse_dates=True)   # header = 1: infer 2nd line 
print(df)
print("---------------------------------------------------------------------------------")

print(df.index)
print("---------------------------------------------------------------------------------")
print(df.tz_localize(tz='US/Eastern'))
print("---------------------------------------------------------------------------------")
print(df)
print("---------------------------------------------------------------------------------")

df.index = df.index.tz_localize(tz='US/Eastern')
print(df.index)
print("---------------------------------------------------------------------------------")

df = df.tz_convert('Europe/Berlin')
print(df)
print("---------------------------------------------------------------------------------")

print(df.index)
```

```python
"""
7.Timezone Handling
"""
import pandas as pd
df = pd.read_csv("fb.csv",parse_dates=['Date'],index_col='Date')
print(df)
print("---------------------------------------------------------------------------------")

print(df.shift(1))
print("---------------------------------------------------------------------------------")
print(df.shift(-1))
print("---------------------------------------------------------------------------------")

df['Prev Day Price'] = df['Price'].shift(1) 
print(df)
print("---------------------------------------------------------------------------------")
df['Price Change'] = df['Price'] - df['Prev Day Price']
print(df)
print("---------------------------------------------------------------------------------")

df['5 day return'] =  (df['Price'] - df['Price'].shift(5))*100/df['Price'].shift(5)
print(df)
print("---------------------------------------------------------------------------------")
df = df[['Price']]
print(df)
print("---------------------------------------------------------------------------------")
print(df.index)
print("---------------------------------------------------------------------------------")
df.index = pd.date_range(start='2017-08-15',periods=10, freq='B')
print(df)
print("---------------------------------------------------------------------------------")
print(df.index)
print("---------------------------------------------------------------------------------")
# Shift the time index, using the indexâs frequency if available.
print(df.tshift(1))
```
